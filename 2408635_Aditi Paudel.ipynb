##  Task- 3- Outlier Detection


```python
#  Identify outlier countries in South Asia based on their Score and GDP per Capita.
Q1_score = df_southasia['score'].quantile(0.25)
Q3_score = df_southasia['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = df_southasia['Log GDP per capita'].quantile(0.25)
Q3_gdp = df_southasia['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Identify outliers based on IQR
outliers_score_iqr = df_southasia[(df_southasia['score'] < Q1_score - 1.5 * IQR_score) | (df_southasia['score'] > Q3_score + 1.5 * IQR_score)]
outliers_gdp_iqr = df_southasia[(df_southasia['Log GDP per capita'] < Q1_gdp - 1.5 * IQR_gdp) | (df_southasia['Log GDP per capita'] > Q3_gdp + 1.5 * IQR_gdp)]

print("Outliers based on IQR for Score:")
print(outliers_score_iqr)

print("\nOutliers based on IQR for GDP per Capita:")
print(outliers_gdp_iqr)
```

```python
#Define outliers using the 1.5  IQR rule.
import pandas as pd
import numpy as np

def detect_outliers(data):
  """Detects outliers in a given dataset using the 1.5 IQR rule.

  Args:
    data: A numpy array or pandas Series containing the data.

  Returns:
    A list of indices corresponding to the outliers.
  """

  q1 = np.percentile(data, 25)
  q3 = np.percentile(data, 75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5 * iqr
  upper_bound = q3 + 1.5 * iqr
  outliers = np.where((data < lower_bound) | (data > upper_bound))[0]
  return outliers

# Assuming you have a DataFrame named 'df_southasia'
outliers_score_indices = detect_outliers(df_southasia['score'])
outliers_gdp_indices = detect_outliers(df_southasia['Log GDP per capita'])

outliers_score = df_southasia.iloc[outliers_score_indices]
outliers_gdp = df_southasia.iloc[outliers_gdp_indices]

print("Outliers based on Score:")
print(outliers_score)

print("\nOutliers based on GDP per Capita:")
print(outliers_gdp)
```

```python
# Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
#in a different color
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'df_southasia'

# Identify outliers using the 1.5 IQR rule
outliers_score_indices = detect_outliers(df_southasia['score'])
outliers_gdp_indices = detect_outliers(df_southasia['Log GDP per capita'])

# Create a scatter plot
plt.figure(figsize=(10, 6))

# Plot all data points in one color
plt.scatter(df_southasia['Log GDP per capita'], df_southasia['score'], color='blue', label='Non-Outliers')

# Plot outliers in a different color
plt.scatter(df_southasia.iloc[outliers_score_indices]['Log GDP per capita'],
            df_southasia.iloc[outliers_score_indices]['score'],
            color='red', label='Outliers (score)')

plt.scatter(df_southasia.iloc[outliers_gdp_indices]['Log GDP per capita'],
            df_southasia.iloc[outliers_gdp_indices]['score'],
            color='red', label='Outliers (Log GDP per capita)')

plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('Scatter Plot of score vs.Log GDP per capita with Outliers')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np

# Function to calculate outliers based on IQR
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)  # 25th percentile
    Q3 = df[column].quantile(0.75)  # 75th percentile
    IQR = Q3 - Q1  # Interquartile range
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# Detect outliers for South Asia
southasia_score_outliers = detect_outliers(df_southasia, 'score')
southasia_gdp_outliers = detect_outliers(df_southasia, 'Log GDP per capita')

# Detect outliers for Middle East
middleeast_score_outliers = detect_outliers(df_middle_east, 'score')
middleeast_gdp_outliers = detect_outliers(df_middle_east, 'Log GDP per capita')

# Scatter plots with outliers highlighted
plt.figure(figsize=(14, 8))

# South Asia
plt.subplot(2, 2, 1)
plt.scatter(df_southasia['Log GDP per capita'], df_southasia['score'], label='Data')
plt.scatter(southasia_gdp_outliers['Log GDP per capita'], southasia_gdp_outliers['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('South Asia: Score vs. GDP per Capita')
plt.legend()

# Middle East
plt.subplot(2, 2, 2)
plt.scatter(df_middle_east['Log GDP per capita'], df_middle_east['score'], label='Data')
plt.scatter(middleeast_gdp_outliers['Log GDP per capita'], middleeast_gdp_outliers['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per capita')
plt.ylabel('score')
plt.title('Middle East: Score vs. GDP per Capita')
plt.legend()

plt.tight_layout()
plt.show()

# Print outlier countries for discussion
print("South Asia Outliers (score):")
print(southasia_score_outliers[['Country name', 'score']])
print("\nSouth Asia Outliers (Log GDP per capita):")
print(southasia_gdp_outliers[['Country name', 'Log GDP per capita']])

print("\nMiddle East Outliers (score):")
print(middleeast_score_outliers[['Country name', 'score']])
print("\nMiddle East Outliers (Log GDP per capita):")
print(middleeast_gdp_outliers[['Country name', 'Log GDP per capita']])
```
