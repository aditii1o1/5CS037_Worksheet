```python
import matplotlib.pyplot as plt  # Import matplotlib for optional plotting

# Temperature readings (one month of data, each day has 3 readings: night, evening, day)
temperatures = [
    8.2, 17.4, 14.1, 7.9, 18.0, 13.5, 9.0, 17.8, 13.0, 8.5,
    16.5, 12.9, 7.7, 17.2, 13.3, 8.4, 16.7, 14.0, 9.5, 18.3,
    13.4, 8.1, 17.9, 14.2, 7.6, 17.0, 12.8, 8.0, 16.8, 13.7,
    7.8, 17.5, 13.6, 8.7, 17.1, 13.8, 9.2, 18.1, 13.9, 8.3,
    16.4, 12.7, 8.9, 18.2, 13.1, 7.8, 16.6, 12.5
]

# Step 1: Create empty lists for night, evening, and day
night_temps = []
evening_temps = []
day_temps = []

# Step 2: Classify temperatures based on their position
for i, temp in enumerate(temperatures):
    if i % 3 == 0:  # Night (positions 0, 3, 6, ...)
        night_temps.append(temp)
    elif i % 3 == 1:  # Evening (positions 1, 4, 7, ...)
        evening_temps.append(temp)
    elif i % 3 == 2:  # Day (positions 2, 5, 8, ...)
        day_temps.append(temp)

# Step 3: Calculate the average day-time temperature
average_day_temp = sum(day_temps) / len(day_temps)
print(f"Average day-time temperature: {average_day_temp:.2f}°C")

# Step 4 (Optional): Plot "day vs. temperature"
days = list(range(1, len(day_temps) + 1))  # Days 1 to the total number of day readings

plt.figure(figsize=(10, 5))
plt.plot(days, day_temps, marker='o', label='Day Temperatures', color='orange')
plt.xlabel("Day")
plt.ylabel("Temperature (°C)")
plt.title("Day vs. Temperature")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def sum_nested_list(nested_list):
    """
    Recursively sums all numbers in a nested list.
    """
    total = 0
    for element in nested_list:
        if isinstance(element, list):  # If the element is a list, recurse
            total += sum_nested_list(element)
        else:  # If the element is a number, add it to the total
            total += element
    return total

# Test the function with a sample nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print(f"The total sum of all numbers in the nested list is: {result}")
```

```python
def generate_permutations(s):
    """
    Recursively generates all unique permutations of a string.
    :param s: The input string
    :return: A list of all unique permutations
    """
    if len(s) == 1:  # Base case: a single character string has one permutation
        return [s]
    
    permutations = []  # List to store all permutations
    for i, char in enumerate(s):
        # Extract the character at index i and find permutations of the rest of the string
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.append(char + perm)  # Add the character back to each permutation

    # Use set to ensure uniqueness, then convert back to a list
    return list(set(permutations))

# Test the function with example strings
print("Permutations of 'abc':", generate_permutations("abc"))
print("Permutations of 'aab':", generate_permutations("aab"))
```

```python
def calculate_directory_size(directory):
    """
    Recursively calculates the total size of a directory.
    :param directory: A dictionary representing the directory structure.
                      Keys are file/subdirectory names, values are file sizes or subdirectories.
    :return: Total size of the directory in KB.
    """
    total_size = 0  # Initialize total size

    for name, content in directory.items():
        if isinstance(content, dict):  # If content is a subdirectory (dictionary)
            total_size += calculate_directory_size(content)  # Recurse into the subdirectory
        else:  # If content is a file (size in KB)
            total_size += content

    return total_size

# Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

# Test the function
total_size = calculate_directory_size(directory_structure)
print(f"Total size of the directory: {total_size} KB")
```

```python
def min_coins(coins, amount):
    """
    Finds the minimum number of coins needed to make up a given amount using dynamic programming.
    :param coins: List of coin denominations
    :param amount: Target amount to make
    :return: Minimum number of coins required to make the amount, or -1 if not possible
    """
    # Initialize the DP array with "infinity" for all amounts, as a placeholder for "not reachable"
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins are needed to make amount 0

    # Iterate over each coin denomination
    for coin in coins:
        # Update the DP array for all amounts that can include the current coin
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    # If the target amount is still "infinity", it means it's not reachable
    return dp[amount] if dp[amount] != float('inf') else -1

# Test the function with the provided example
coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"The minimum number of coins needed to make {amount} is: {result}")
```

```python
def longest_common_subsequence(s1, s2):
    """
    Finds the length of the longest common subsequence (LCS) between two strings using dynamic programming.
    :param s1: First string
    :param s2: Second string
    :return: Length of the LCS
    """
    # Initialize a 2D DP table with dimensions (len(s1) + 1) x (len(s2) + 1)
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]

    # Fill the DP table
    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:  # If characters match
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:  # Otherwise, take the maximum from previous values
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    # The bottom-right cell contains the length of the LCS
    return dp[len(s1)][len(s2)]

# Test the function with example strings
s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"The length of the longest common subsequence between '{s1}' and '{s2}' is: {result}")
```
